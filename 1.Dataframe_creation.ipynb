### 2.1) Key_words by day with clasification
Stock the dictionary of all the words with the count value.    

RangeIndex: 838 entries, 0 to 837   
Data columns (total 9 columns):
* year          -   int   
* month         -   int   
* day           -   int   
* kws           -   dict 
* locations     -   dict 
* miscellaneus  -   dict 
* organizations -   dict 
* persons       -   dict 
* num_articles  -   int   

```python
df_key_words = pd.DataFrame(columns=["year", "month", "day", "kws", "locations", "miscellaneus", "organizations", "persons", "num_articles"])
```

```python
Years = ["2019", "2020", "2021"]

for year in Years:
    for month in range(1, 13): 
        for day in range(1, 32): 
            try :
                data_tmp = data['metadata-all']['fr']['day'][str(year)][str(month)][str(day)]
                new_data = pd.DataFrame({"year": year, 
                                        'month': month,
                                        'day': day,
                                        'kws': [data_tmp["kws"]],
                                        'locations': [data_tmp["loc"]],
                                        'miscellaneus': [data_tmp["mis"]],
                                        'organizations': [data_tmp["org"]],
                                        'persons': [data_tmp["per"]], 
                                        'num_articles': data_tmp["num"]
                                        })
                df_key_words = pd.concat([df_key_words, new_data], ignore_index=True)
            except:
                continue
```

### 2.1) Key_words by article

```python
df_kws_article = pd.DataFrame(columns=["year", "month", "day", "url", "title", "description", "content", "lang", "kws", "locations", "miscellaneus", "organizations", "persons"])
```

```python
Years = ["2019", "2020", "2021"]

for year in Years:
    for month in range(1, 13): 
        for day in range(1, 32): 
            try:
                num = 0
                while(True):
                    try :
                        data_tmp = data['data-all'][str(year)][str(month)][str(day)][num]
                        num += 1
                        
                        new_data = pd.DataFrame({"year": year, 
                                                'month': month,
                                                'day': day,
                                                'url' : data_tmp["url"],
                                                'title' : data_tmp["title"],
                                                'description' : data_tmp["description"],
                                                'content' : data_tmp["content"],
                                                 'lang' : data_tmp["lang"],
                                                'kws': [data_tmp["kws"]],
                                                'locations': [data_tmp["loc"]],
                                                'miscellaneus': [data_tmp["mis"]],
                                                'organizations': [data_tmp["org"]],
                                                'persons': [data_tmp["per"]]
                                                })

                        df_kws_article = pd.concat([df_kws_article, new_data], ignore_index=True)
                    except:
                        break
            except:
                continue
```

### 2.3) Key_words multiple articles

```python
df_multiple_articles = pd.DataFrame(columns=["year", "month", "day", "url", "title", "description", "content", "lang", "kws", "locations", "miscellaneus", "organizations", "persons"])
```

```python
file_paths = [
    "./preprocess-topaz-data732/topaz-data732--france--fr.sputniknews.africa--20190101--20211231.json",
    "./preprocess-topaz-data732/topaz-data732--france--www.egaliteetreconciliation.fr--20190101--20211231.json",
    "./preprocess-topaz-data732/topaz-data732--france--french.presstv.ir--20190101--20211231.json",
    "./preprocess-topaz-data732/topaz-data732--france--www.fdesouche.com--20190101--20211231.json",
    "./preprocess-topaz-data732/topaz-data732--mali--french.presstv.ir--20190101--20211231.json",
    "./preprocess-topaz-data732/topaz-data732--mali--fr.sputniknews.africa--20190101--20211231.json",
    "./preprocess-topaz-data732/topaz-data732--mali--www.egaliteetreconciliation.fr--20190101--20211231.json"
]
```

```python
def get_pd_kws(Data):
    Years = ["2019", "2020", "2021"]
    df_final = pd.DataFrame(columns=["year", "month", "day", "url", "title", "description", "content", "lang", "kws", "locations", "miscellaneus", "organizations", "persons"]) 
    for year in Years:
        for month in range(1, 13): 
            for day in range(1, 32): 
                try:
                    num = 0
                    while(True):
                        try :
                            data_tmp = Data['data-all'][str(year)][str(month)][str(day)][num]
                            num += 1

                            new_data = pd.DataFrame({"year": year, 
                                                    'month': month,
                                                    'day': day,
                                                    'url' : data_tmp["url"],
                                                    'title' : data_tmp["title"],
                                                    'description' : data_tmp["description"],
                                                    'content' : data_tmp["content"],
                                                     'lang' : data_tmp["lang"],
                                                    'kws': [data_tmp["kws"]],
                                                    'locations': [data_tmp["loc"]],
                                                    'miscellaneus': [data_tmp["mis"]],
                                                    'organizations': [data_tmp["org"]],
                                                    'persons': [data_tmp["per"]]
                                                    })

                            df_final = pd.concat([df_final, new_data], ignore_index=True)
                        except:
                            break
                except:
                    continue
    return df_final
```

```python

for path in file_paths:
    try:
        with open(path, "r") as json_file:
            data_tmp = json.load(json_file)
    except FileNotFoundError:
        print(f"File not found: {file_path}")
    except json.JSONDecodeError as e:
        print(f"JSON decoding error: {e}")
        
    df_multiple_articles = pd.concat([df_multiple_articles, get_pd_kws(data_tmp)], ignore_index=True)
```
